```python
import xml.etree.cElementTree as ET
import lxml.etree
from xml.etree.ElementTree import Element, SubElement, ElementTree
from xml.dom.minidom import parseString


def text_modification(root):
    for review in root.iter("num_of_reviews"):
        if int(review.text) < 1000:
            review.text = str(0)
    for author in root.iter('author'):
        author.text = author.text.replace('By ', '')
    for rating in root.iter('rating'):
        rating.text = str(round(float(rating.text), 2))
    return root


def attribute_modification(root):
    for recipe in root.iter('recipe'):
        recipe.attrib['id'] = 'recipe_' + recipe.attrib['id']
    for rating in root.iter('rating'):
        rating.set('max_value', '5.00')
    for author in root.iter('author'):
        author.set('value', 'username')
    return root


def elements_adding(root):
    recipe = SubElement(root, 'recipe')
    recipe.set('id', 'recipe_' + '20')
    name = SubElement(recipe, 'name')
    name.text = 'Ratatouille'
    num_of_reviews = SubElement(recipe, 'num_of_reviews')
    num_of_reviews.text = '1675'
    rating = SubElement(recipe, 'rating')
    rating.set('max_value', '5.00')
    rating.text = '4.85'
    description = SubElement(recipe, 'description')
    description.text = 'National French dish...'
    author = SubElement(recipe, 'author')
    author.set('value', 'username')
    author.text = 'Neichev'
    return root


def create_html():
    dom = lxml.etree.parse('update_recipe.xml')
    xslt = lxml.etree.parse('update_recipe.xslt')
    transform = lxml.etree.XSLT(xslt)
    newdom = transform(dom)
    with open('update_recipe.html', 'w', encoding='utf-8') as file:
        file.write(str(newdom))
        file.close()


if __name__ == '__main__':
    tree = ET.ElementTree(file="recipes.xml")
    root = tree.getroot()
    root = text_modification(root)
    root = attribute_modification(root)
    root = elements_adding(root)
    tree = ET.ElementTree(root)
    create_html()
```
